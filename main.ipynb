```python
# 1. compute Age from Year_Birth
# 2. parse Dt_Customer and compute tenure (CustomerDays)
# 3. compute TotalMnt and TotalPurchases
# 4. compute ChildrenCount

df2 = df.copy()

# Date parsing (coerce invalid)
df2["Dt_Customer"] = pd.to_datetime(df2["Dt_Customer"], errors="coerce")

# Age relative to 2025 (use absolute year for reproducibility)
df2["Age"] = 2025 - df2["Year_Birth"]

# Monetary and purchase aggregates
monetary = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
for c in monetary:
    if c not in df2.columns:
        raise KeyError(f"Expected column '{c}' not found in dataset")

df2["TotalMnt"] = df2[monetary].sum(axis=1)

purchase_cols = ["NumDealsPurchases", "NumWebPurchases", "NumStorePurchases", "NumCatalogPurchases"]
for c in purchase_cols:
    if c not in df2.columns:
        raise KeyError(f"Expected column '{c}' not found in dataset")

df2["TotalPurchases"] = df2[purchase_cols].sum(axis=1)

# Customer tenure in days (use a fixed 'now' date; change if you want current date)
NOW = pd.to_datetime("2025-09-22")
df2["CustomerDays"] = (NOW - df2["Dt_Customer"]).dt.days

# children count
df2["ChildrenCount"] = df2.get("Kidhome", 0) + df2.get("Teenhome", 0)

# Fill / impute Income if missing (median)
df2["Income"] = df2["Income"].fillna(df2["Income"].median())

# Inspect engineered columns
display(df2[["ID","Year_Birth","Age","Income","TotalMnt","TotalPurchases","CustomerDays","ChildrenCount"]].head())
```

```python
# Outlier filtering
print("Rows before filtering:", df2.shape[0])
# - simple outlier handling: remove rows with unrealistic age or negative tenure
# - optionally you can use more advanced outlier detection later

# Filter unrealistic ages
df2 = df2[(df2["Age"] >= 18) & (df2["Age"] <= 100)]

# Filter rows with negative or NaN CustomerDays
df2 = df2[df2["CustomerDays"].notna() & (df2["CustomerDays"] >= 0)]

# Reset index
df2 = df2.reset_index(drop=True)
print("Rows after filtering:", df2.shape[0])
```

```python
# Cell 4: Feature selection and scaling
# We'll pick numeric features that are meaningful for customer segmentation
# You can add / remove features depending on business focus.

from sklearn.preprocessing import StandardScaler

features = [
    "Age",
    "Income",
    "TotalMnt",
    "TotalPurchases",
    "Recency",             # how recent was last purchase
    "NumWebVisitsMonth",   # engagement proxy
    "ChildrenCount",
    "CustomerDays"         # tenure
]

# confirm these columns exist (some datasets vary)
missing = [c for c in features if c not in df2.columns]
if missing:
    raise KeyError(f"Missing expected feature columns: {missing}")

X = df2[features].copy()

# scale for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# keep X and scaler for inverse transform later
print("Features used for clustering:", features)
display(pd.DataFrame(X_scaled, columns=features).head())
```

```python
# 5: Elbow (WCSS) and Silhouette diagnostics
# - compute inertia (WCSS) and silhouette scores for k=2..8
# - visualize both plots (two separate plots)
# - clustering is relatively cheap for this dataset size; if your environment is limited,
# you can reduce the k-range.

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

Ks = list(range(2, 9))
wcss = []
sil_scores = []

for k in Ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    lbl = km.fit_predict(X_scaled)
    wcss.append(km.inertia_)
    sil_scores.append(silhouette_score(X_scaled, lbl))

# Plot WCSS (Elbow)
plt.figure(figsize=(7,4))
plt.plot(Ks, wcss, marker="o")
plt.xlabel("k (number of clusters)")
plt.ylabel("WCSS (inertia)")
plt.title("Elbow method (WCSS)")
plt.grid(True)
plt.show()

# Plot silhouette
plt.figure(figsize=(7,4))
plt.plot(Ks, sil_scores, marker="o")
plt.xlabel("k (number of clusters)")
plt.ylabel("Silhouette score")
plt.title("Silhouette scores by k")
plt.grid(True)
plt.show()

# Print values
for k, s in zip(Ks, sil_scores):
    print(f"k={k} -> silhouette: {s:.4f}")
```

```python
# Cell 6: Fit final K-Means and (optional) GMM
# Pick k based on diagnostics above (e.g., best silhouette or elbow). 
# Here we programmatically pick the best k by max silhouette.

from sklearn.mixture import GaussianMixture

best_k = Ks[np.argmax(sil_scores)]
print("Automatically selected best_k (by silhouette):", best_k)

# Fit KMeans
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=50)
klabels = kmeans.fit_predict(X_scaled)

# Optionally fit GMM for soft clustering / comparison
gmm = GaussianMixture(n_components=best_k, random_state=42)
gmm_labels = gmm.fit_predict(X_scaled)

# attach labels to df2
df2["KMeansCluster"] = klabels
df2["GMMCluster"] = gmm_labels

print("Cluster counts (KMeans):")
print(df2["KMeansCluster"].value_counts().sort_index())
```

```python
# 7: Cluster profiling (summary statistics and interpretation)
# - show mean of key features per cluster (in original units)
# - also show counts and some percentiles for TotalMnt

# Prepare a DataFrame of cluster summaries (using original scale)
cluster_summary = df2.groupby("KMeansCluster")[features].agg(["mean","median","std","count"])
# Flatten columns
cluster_summary.columns = ["_".join(col).strip() for col in cluster_summary.columns.values]

# Add average TotalMnt and TotalPurchases per cluster for quick business interpretation
cluster_stats = df2.groupby("KMeansCluster").agg(
    Count = ("ID","count"),
    Avg_Age = ("Age","mean"),
    Avg_Income = ("Income","mean"),
    Avg_TotalMnt = ("TotalMnt","mean"),
    Median_TotalMnt = ("TotalMnt","median"),
    Avg_TotalPurchases = ("TotalPurchases","mean"),
    Avg_NumWebVisitsMonth = ("NumWebVisitsMonth","mean"),
    Avg_ChildrenCount = ("ChildrenCount","mean")
).sort_index()

display(cluster_stats)
# Show more detailed summary
display(cluster_summary.T)
```

```python
# Cell 8: PCA for 2D visualization of clusters
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(7,5))
for c in sorted(df2["KMeansCluster"].unique()):
    mask = df2["KMeansCluster"] == c
    plt.scatter(X_pca[mask,0], X_pca[mask,1], label=f"Cluster {c}", alpha=0.6, s=40)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("KMeans clusters visualized on PCA components")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cell 9: Boxplots and simple visual comparisons
# Example: TotalMnt distribution by cluster

plt.figure(figsize=(8,5))
df2.boxplot(column="TotalMnt", by="KMeansCluster", grid=True)
plt.suptitle("")   # hide the default suptitle
plt.title("Total monetary spend by KMeans cluster")
plt.xlabel("KMeans cluster")
plt.ylabel("TotalMnt")
plt.show()

# Bar chart: average income per cluster
plt.figure(figsize=(7,4))
cluster_stats["Avg_Income"].plot(kind="bar")
plt.title("Average Income per cluster")
plt.xlabel("KMeans cluster")
plt.ylabel("Average Income")
plt.show()
```

```python
# Cell X2: Gaussian Mixture Model (GMM) + visualization
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=best_k, random_state=42)
glabels = gmm.fit_predict(X_scaled)

plt.figure(figsize=(7,5))
for c in np.unique(glabels):
    plt.scatter(X_pca[glabels==c,0], X_pca[glabels==c,1], label=f"Cluster {c}", s=40, alpha=0.6)
plt.title("GMM Clusters (PCA projection)")
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.legend(); plt.grid(True)
plt.show()
```

```python
# Cell X3: DBSCAN clustering + visualization
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.5, min_samples=10)  # tune eps & min_samples
dlabels = dbscan.fit_predict(X_scaled)

plt.figure(figsize=(7,5))
unique_labels = np.unique(dlabels)
for c in unique_labels:
    label = f"Cluster {c}" if c != -1 else "Noise"
    plt.scatter(X_pca[dlabels==c,0], X_pca[dlabels==c,1], label=label, s=40, alpha=0.6)
plt.title("DBSCAN Clusters (PCA projection)")
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.legend(); plt.grid(True)
plt.show()

print("DBSCAN produced clusters:", np.unique(dlabels))
```

```python
# Cell X4: Agglomerative Hierarchical clustering + visualization
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=best_k, linkage="ward")
alabels = agg.fit_predict(X_scaled)

plt.figure(figsize=(7,5))
for c in np.unique(alabels):
    plt.scatter(X_pca[alabels==c,0], X_pca[alabels==c,1], label=f"Cluster {c}", s=40, alpha=0.6)
plt.title("Hierarchical Clusters (PCA projection)")
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.legend(); plt.grid(True)
plt.show()
```

```python
# Cell X5: Dendrogram for Hierarchical Clustering
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(12,6))
# linkage matrix using Ward's method (good for Euclidean distances)
linkage_matrix = sch.linkage(X_scaled, method="ward")

# dendrogram plot
dendro = sch.dendrogram(
    linkage_matrix,
    truncate_mode="level",   # show only top levels (cleaner)
    p=20,                    # number of leaf nodes shown
    leaf_rotation=45,
    leaf_font_size=10,
    show_contracted=True
)

plt.title("Hierarchical Clustering Dendrogram (truncated)")
plt.xlabel("Sample index or (cluster size)")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.cluster import AgglomerativeClustering, DBSCAN

results = {}

# --- KMeans ---
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=50)
kmeans_labels = kmeans.fit_predict(X_scaled)
results["KMeans"] = [
    silhouette_score(X_scaled, kmeans_labels),
    davies_bouldin_score(X_scaled, kmeans_labels),
    calinski_harabasz_score(X_scaled, kmeans_labels)
]

# --- GMM ---
gmm = GaussianMixture(n_components=best_k, random_state=42)
gmm_labels = gmm.fit_predict(X_scaled)
results["GMM"] = [
    silhouette_score(X_scaled, gmm_labels),
    davies_bouldin_score(X_scaled, gmm_labels),
    calinski_harabasz_score(X_scaled, gmm_labels)
]

# --- DBSCAN ---
dbscan = DBSCAN(eps=1.8, min_samples=8)  # tweak eps if needed
dbscan_labels = dbscan.fit_predict(X_scaled)

# DBSCAN sometimes assigns all points to -1 (noise), so check
if len(set(dbscan_labels)) > 1 and len(set(dbscan_labels)) < len(X_scaled):
    results["DBSCAN"] = [
        silhouette_score(X_scaled, dbscan_labels),
        davies_bouldin_score(X_scaled, dbscan_labels),
        calinski_harabasz_score(X_scaled, dbscan_labels)
    ]
else:
    results["DBSCAN"] = ["Invalid", "Invalid", "Invalid"]

# --- Hierarchical ---
agg = AgglomerativeClustering(n_clusters=best_k, linkage="ward")
agg_labels = agg.fit_predict(X_scaled)
results["Hierarchical"] = [
    silhouette_score(X_scaled, agg_labels),
    davies_bouldin_score(X_scaled, agg_labels),
    calinski_harabasz_score(X_scaled, agg_labels)
]

# --- Show results ---
metrics = ["Silhouette", "Davies-Bouldin", "Calinski-Harabasz"]
results_df = pd.DataFrame(results, index=metrics).T
display(results_df)
```

### **Interpretation**

**KMeans**: Good compactness (highest CHI), but clusters overlap â†’ silhouette low.

**GMM**: Performed worst overall â†’ discard.

**DBSCAN**: Found some structure (high silhouette) but essentially threw most points into noise or 1 cluster (low CHI).

**Hierarchical**: More balanced, best DBI (clusters least overlapping), slightly behind KMeans in CHI.

### **Recommendation**

KMeans is the most practical choice for segmentation (good balance + interpretable).

Hierarchical supports the KMeans result (consistent clustering structure).

DBSCAN is not really suitable here.

GMM is not effective â†’ discard.
